```python
import polars as pl

df_initial = pl.read_csv("merged3_with_year_built.csv")

df_initial
```

```python
df_dropped_column  = df_initial.drop(["Year_built", "Borough", "LSOA 2011 Name", "LA Name", "2012", "2013", "2014", "2015", "2016", "LAD Code", "LA Name", "Total Mean Annual Household Income estimate"])
df_dropped_column
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np
from sklearn.metrics import mean_squared_error, r2_score

x = df_dropped_column["Total Median Annual Household Income estimate"].to_numpy().reshape(-1, 1)  # X should be a 2D array
y = df_dropped_column["Energy Efficiency"].to_numpy()  # y is a 1D array

# Perform linear regression using sklearn
model = LinearRegression()
model.fit(x, y)

# Print regression results
print("Intercept:", model.intercept_)
print("Coefficient:", model.coef_[0])

# Use the model to make predictions
predictions = model.predict(x)
print("Predictions:", predictions)

model

correlation_matrix = np.corrcoef(df_dropped_column["Total Median Annual Household Income estimate"].to_numpy(), df_dropped_column["Energy Efficiency"].to_numpy())
pearson_coefficient = correlation_matrix[0, 1]
print("Pearson Correlation Coefficient (r):", pearson_coefficient)

# Calculate R² (Coefficient of Determination)
r2 = r2_score(y, predictions)
print("R² (Coefficient of Determination):", r2)

# Calculate Mean Squared Error (MSE)
mse = mean_squared_error(y, predictions)
print("Mean Squared Error (MSE):", mse)
```

```python
import matplotlib.pyplot as plt

plt.scatter(x, y, color="blue", label="Actual Data (2017 vs Energy Efficiency)")
plt.plot(x, predictions, color="red", label="Regression Line")

# Add labels, title, and legend to the plot
plt.xlabel("Total Median Annual Household Income estimate")
plt.ylabel("Energy Efficiency")
plt.title("Linear Regression on Energy Efficiency Data")
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```
